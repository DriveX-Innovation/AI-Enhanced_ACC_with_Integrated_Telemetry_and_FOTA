```python
import sys
carla_egg = r"D:\graduation project related\Vehicle-CV-ADAS-master\WindowsNoEditor\PythonAPI\carla\dist\carla-0.9.11-py3.7-win-amd64.egg"

if carla_egg not in sys.path:
    sys.path.append(carla_egg)

import carla
```

```python


def main():
    try:
        client = carla.Client('localhost', 2000)
        client.set_timeout(10.0)
        world = client.get_world()

        # 1. Get all environment objects tagged as 'Buildings'
        # You can also add other tags like Fences or Walls if needed
        print("Searching for building objects...")
        env_objs = world.get_environment_objects(carla.CityObjectLabel.Buildings)

        # 2. Extract the unique IDs of these objects
        building_ids = [obj.id for obj in env_objs]

        if not building_ids:
            print("No building objects found! (Are you on a custom map without tags?)")
            return

        # 3. Disable the objects (Removes visuals and physics)
        print(f"Found {len(building_ids)} building objects. Removing them...")
        world.enable_environment_objects(building_ids, False) 
        
        print("Buildings removed successfully.")

    except Exception as e:
        print(f"Error: {e}")

if __name__ == '__main__':
    main()
```
